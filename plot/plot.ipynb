```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import ticker
```

```python
df = pd.read_csv('../utils/train.csv')

plt.figure(figsize=(10, 6))
plt.plot(df['Epoch'], df['Train_loss'], label='Train Loss', color='blue')

plt.title('Training Loss Over Epochs')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()

plt.savefig('loss_plot.png')
```

```python
df = pd.read_csv('../utils/train.csv')

plt.figure(figsize=(10, 6))
plt.plot(df['Epoch'], df['Test_acc']*100, label='Test Acc', color='red')

max_acc_idx = df['Test_acc'].idxmax()
max_acc = df['Test_acc'].iloc[max_acc_idx] * 100
max_acc_epoch = df['Epoch'].iloc[max_acc_idx]

plt.scatter(max_acc_epoch, max_acc, color='blue', s=100, zorder=5, label=f'Max Test Acc ({max_acc:.2f}%)')

plt.title('Test Acc Over Epochs')
plt.xlabel('Epoch')
plt.ylabel('Test Acc')

plt.gca().yaxis.set_major_formatter(ticker.PercentFormatter())
plt.legend()

plt.savefig('acc_plot.png')
```
